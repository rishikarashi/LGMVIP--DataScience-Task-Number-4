### IMPORTING THE LIBRARIES:

```python
import cv2
import matplotlib.pyplot as plt
```

### IMPORTING THE IMAGE IN RGB FORMAT:

```python
image= 'C:\\Users\\Rishika Rashi\\Downloads\\FLOWER.jpg'
flower=cv2.imread(image)
flower_rgb=cv2.cvtColor(flower, cv2.COLOR_BGR2RGB)
plt.imshow(flower_rgb)
plt.axis('off')
plt.show()
```

### CONVERTING THE IMAGE IN GRAYSCALE

```python
gray_flower = cv2.cvtColor(flower, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_flower, cmap='gray')
plt.axis('off')
plt.show()
```

### INVERTING THE GRAYSCALE IMAGE

```python
inverted_flower = 255 - gray_flower
plt.imshow(inverted_flower, cmap='gray')
plt.axis('off')
plt.show()
```

### CREATING THE PENCIL SKETCH

```python
pencil_sketch = cv2.divide(gray_flower, inverted_flower, scale=300.0)
plt.imshow(pencil_sketch, cmap='gray')
plt.axis('off')
plt.show()
```

### DISPLAYING ALL THE IMAGES:

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].imshow(flower_rgb)
axs[0, 0].axis('off')
axs[0, 0].set_title('Original Image')
axs[0, 1].imshow(gray_flower, cmap='gray')
axs[0, 1].axis('off')
axs[0, 1].set_title('Grayscale Image')
axs[1, 0].imshow(inverted_flower, cmap='gray')
axs[1, 0].axis('off')
axs[1, 0].set_title('Inverted Image')
axs[1, 1].imshow(pencil_sketch, cmap='gray')
axs[1, 1].axis('off')
axs[1, 1].set_title('Pencil Sketch')
plt.tight_layout()
plt.show()
```
